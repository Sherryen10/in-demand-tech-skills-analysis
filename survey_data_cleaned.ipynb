```python
# Prepare DataFrame only which columns required and check info
df_cleaned = df[['ResponseId', 'Age', 'EdLevel', 'Country', 'LanguageHaveWorkedWith',
          'DatabaseHaveWorkedWith', 'PlatformHaveWorkedWith',
          'WebframeHaveWorkedWith', 'LanguageWantToWorkWith',
          'DatabaseWantToWorkWith', 'PlatformWantToWorkWith',
          'WebframeWantToWorkWith']].copy()
df_cleaned.info()
```

```python
# Map messy education levels to simplified categories
edlevel_mapping = {
    "Bachelor's degree (B.A., B.S., B.Eng., etc.)": "Bachelor",
    "Master's degree (M.A., M.S., M.Eng., MBA, etc.)": "Master",
    "Professional degree (JD, MD, Ph.D, Ed.D, etc.)": "Doctorate",
    "Associate degree (A.A., A.S., etc.)": "Associate",
    "Some college/university study without earning a degree": "Some College",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "Secondary School",
    "Primary/elementary school": "Primary School",
    "Something else": "Other"
}

df_cleaned['EdLevel'] = df_cleaned['EdLevel'].map(edlevel_mapping)
display(df_cleaned['EdLevel'].unique())
```

```python
# Check Country column
df_cleaned['Country'].unique()
```

```python
# Standardized Country name
country_mapping = {
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'United States of America': 'United States',
    'Republic of North Macedonia': 'North Macedonia',
    'Russian Federation': 'Russia',
    'United Republic of Tanzania': 'Tanzania',
    'Iran, Islamic Republic of...': 'Iran',
    'Venezuela, Bolivarian Republic of...': 'Venezuela',
    'Hong Kong (S.A.R.)': 'Hong Kong',
    'Democratic Republic of the Congo': 'DR Congo',
    'Congo, Republic of the...': 'Congo (Republic)',
    'Viet Nam': 'Vietnam',
    'Republic of Korea': 'South Korea',
    'Syrian Arab Republic': 'Syria'
}

df_cleaned['Country'] = df_cleaned['Country'].replace(country_mapping)
display(df_cleaned['Country'].unique())
```

```python
# Define multi-item columns
multi_item_columns = [
    'LanguageHaveWorkedWith',
    'DatabaseHaveWorkedWith',
    'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'LanguageWantToWorkWith',
    'DatabaseWantToWorkWith',
    'PlatformWantToWorkWith',
    'WebframeWantToWorkWith'
]
```

```python
# Create reusable normalize function
def normalize_multivalue_column(df, column_name, id_col='ResponseId'):
    """
    Splits a semicolon-separated column into one row per value.
    """
    temp = df[[id_col, column_name]].dropna().copy()
    temp[column_name] = temp[column_name].str.split(';')
    temp = temp.explode(column_name)
    temp[column_name] = temp[column_name].str.strip()
    temp = temp[temp[column_name] != '']  # remove empty values
    return temp
```

```python
# Normalize all multi-item columns
normalized_tables = {}

for col in multi_item_columns:
    normalized_tables[col] = normalize_multivalue_column(df_cleaned, col)
```

```python
# Export all columns to csv file
for col, df_norm in normalized_tables.items():
    output_name = f"survey_{col.lower()}_normalized.csv"
    df_norm.to_csv(output_name, index=False)
    print(f"Saved: {output_name}")
```

```python
# Export all demographic columns to csv file
demographic_columns = ['ResponseId', 'Age', 'EdLevel', 'Country']
df_demographics = df_cleaned[demographic_columns].copy()
df_demographics.to_csv('survey_demographics.csv', index=False)
```
